# Importing important libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Opening the file

```python
data = pd.read_csv('data_jobs.csv')
```

```python
print(data)
```

```python
df = pd.DataFrame(data)
```

```python
df.info()
```

# Cleaning

```python
# changing the column's data type into a suitable one (datetime)
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df.info()
```

```python
df
```

```python
# adding a month column
df['hob_posted_month'] = df.job_posted_date.dt.month
```

```python
df.info()
```

```python
df.rename(columns= {'hob_posted_month': 'job_posted_month'}, inplace= True)
```

```python
df.info()
```

```python
# sorting by date
df.sort_values(by='job_posted_date', inplace= True)
```

```python
df
```

```python
# dropping unnecessary columns (salary_hour_avg) 
df.drop(labels= 'salary_hour_avg', axis= 1, inplace= True)
```

```python
df.info()
```

```python
# dropping Null values
df.dropna(subset=['salary_year_avg'], inplace= True)
```

```python
# reseting index to avoid index out of range problem
df.reset_index(drop=True, inplace=True)
```

```python
df.info()
```

# Visualization

```python
month_counts = df['job_posted_month'].value_counts()
month_counts = month_counts.sort_index()
```

```python
plt.plot(month_counts.index, month_counts)
```

```python
job_counts = df['job_title_short'].value_counts()
```

```python
job_counts = job_counts.sort_values(ascending= True)
plt.barh(job_counts.index, job_counts)
plt.title('Posting Jobs')
plt.ylabel('Job Title')
plt.xlabel('Count Of Jobs')
plt.show()
```

```python
job_sal = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
job_sal.plot(kind='barh')
plt.xlabel('Salary $USD')
plt.ylabel('')
plt.title('Medien Slary By Job Title')
plt.show()
```
